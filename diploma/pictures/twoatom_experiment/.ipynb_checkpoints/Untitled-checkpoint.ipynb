```python
%matplotlib notebook

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math as m

# matplotlib plotting parameters
mpl.rcParams['text.usetex'] = True
#mpl.rcParams['text.latex.unicode'] = True

mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Times'
mpl.rcParams['figure.titlesize'] = 'xx-large'
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['axes.titlesize'] = 'xx-large'
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12

# подключаем пакеты для рендеринга русского текста в LaTeX
mpl.rc('text.latex', preamble=r'\usepackage{amsmath}')
mpl.rc('text.latex', preamble=r"\usepackage[utf8]{inputenc}")
mpl.rc('text.latex', preamble=r"\usepackage[russian]{babel}")

from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
ne_ar = np.loadtxt("./ne_ar.txt", delimiter=',')
ar_kr = np.loadtxt("./ar_kr.txt", delimiter=',')
he_ar = np.loadtxt("./he_ar.txt", delimiter=',')

ne_ar[:,1] *= 1e7
ar_kr[:,1] *= 1e7
he_ar[:,1] *= 1e7
```

```python
fig = plt.figure(figsize=(7.5, 6))
host = plt.subplot(111)

host.scatter( ne_ar[:,0], ne_ar[:,1], color = 'k', marker = 'x', s = 15, linewidth = 0.75 )
host.scatter( ar_kr[:,0], ar_kr[:,1], color = 'k', marker = '^', s = 9, linewidth = 0.75 )
host.scatter( he_ar[:,0], he_ar[:,1], color = 'k', marker = 'o', s = 7, linewidth = 0.75 )

plt.xlim((0.0, 450.0))
plt.ylim((0.0, 8.0))

legend_elements = [
    Line2D([0], [0], linewidth = 0.0, marker = 'x', color = 'k', label = 'Ne$-$Ar'),
    Line2D([0], [0], linewidth = 0.0, marker = '^', color = 'k', label = 'Ar$-$Kr'),
    Line2D([0], [0], linewidth = 0.0, marker = 'o', color = 'k', label = 'He$-$Ar'),
] 

host.set_xlabel("Частота, см$^{-1}$")
plt.ylabel(r'$\alpha \cdot 10^7 / \rho_1 \rho_2$ (см$^{-1} \cdot$ Амага$^{-2}$)')

host.legend( handles = legend_elements, prop = {'size': 12}, loc = 'upper right' )

plt.plot()
```
